### 1.1 

Create a function to calculate the probability density function (PDF) of  beta distribution:

\begin{equation*}
f(x;\alpha,\beta) = \frac{\Gamma(\alpha+\beta)}{\Gamma(\alpha)\Gamma(\beta)}x^{\alpha-1}(1-x)^{\beta-1}
\end{equation*}


* Function name: pdf_beta

* Input: $\alpha$, $\beta$, x

* Output: f(x;$\alpha$,$\beta$)

* Use SpecialFunctions.gamma() for the gamma function


```julia
import Pkg
Pkg.add(["SpecialFunctions", "Plots"])
```

```julia
import SpecialFunctions
import Plots
```

```julia
function pdf_beta(α, β, x)
    first_term = SpecialFunctions.gamma(α+β) / (SpecialFunctions.gamma(α)*SpecialFunctions.gamma(β))
    return first_term * x.^(α - 1) .* (1 .- x).^(β - 1)
end
```

According to Bayes' theorem, we have:

\begin{equation*}
Posterior = \frac{Likelihood * Prior}{Marginal}
\end{equation*}

Since we assume no influence from marginal probability, we can ignore the denominator and substitute our other values...

\begin{equation*}
P(Hypothesis|Data) \propto {P(Data|Hypothesis)*P(Hypothesis)}
\end{equation*}

Given:

\begin{equation*}
P(Data|Hypothesis) \sim N(\mu_1, \sigma_1)
\end{equation*}

\begin{equation*}
P(Hypothesis) \sim N(\mu_2, \sigma_2)
\end{equation*}

\begin{equation*}
N(\mu, \sigma) \sim \frac{1}{\sqrt{2\pi\sigma^2}}e^{-\frac{1}{2}(\frac{x-\mu}{\sigma})^2}
\end{equation*}

Substituting the likelihood and prior into the Bayes' theorem, we get:

\begin{equation*}
P(Hypothesis|Data) \propto \frac{1}{\sqrt{2\pi\sigma_1^2}}e^{-\frac{1}{2}(\frac{x-\mu_1}{\sigma_1})^2}\frac{1}{\sqrt{2\pi\sigma_2^2}}e^{-\frac{1}{2}(\frac{x-\mu_2}{\sigma_2})^2}
\end{equation*}

After factoring and simplification...

\begin{equation*}
P(Hypothesis|Data) \propto \frac{1}{2\pi\sigma_1\sigma_2}e^{-\frac{1}{2}\left[\frac{(x-\mu_1)^2}{\sigma_1^2}+\frac{(x-\mu_2)^2}{\sigma_2^2}\right]}
\end{equation*}

If we wish to learn the value of mu, let's consider the left hand side...

\begin{equation*}
P(Hypothesis|Data) \sim N(\mu_{posterior}, \sigma_{posterior})
\end{equation*}

The updated mean and variance is found through mean and variance of the posterior/Gaussian distribution:

\begin{equation*}
\mu_{posterior} = \frac{\frac{\mu_1}{\sigma_1^2} + \frac{\mu_2}{\sigma_2^2}}{\frac{1}{\sigma_1^2} + \frac{1}{\sigma_2^2}} = \frac{\mu_1\sigma_2^2 + \mu_2\sigma_1^2}{\sigma_1^2 + \sigma_2^2}
\end{equation*}

\begin{equation*}
\sigma_{posterior}^2 = \left(\frac{1}{\sigma_1^2} + \frac{1}{\sigma_2^2}\right)^{-1} = \frac{\sigma_1^2\sigma_2^2}{\sigma_1^2 + \sigma_2^2}
\end{equation*}
