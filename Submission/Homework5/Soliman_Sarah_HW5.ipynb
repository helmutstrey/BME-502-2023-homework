### 1.1 

Create a function to calculate the probability density function (PDF) of  beta distribution:

\begin{equation*}
f(x;\alpha,\beta) = \frac{\Gamma(\alpha+\beta)}{\Gamma(\alpha)\Gamma(\beta)}x^{\alpha-1}(1-x)^{\beta-1}
\end{equation*}


* Function name: pdf_beta

* Input: $\alpha$, $\beta$, x

* Output: f(x;$\alpha$,$\beta$)

* Use SpecialFunctions.gamma() for the gamma function


```julia
import Pkg
Pkg.add(["SpecialFunctions", "Plots"])
```

```julia
import SpecialFunctions as sp
import Plots
```

```julia
function pdf_beta(α,β,x::Number)
    #create a variable for gamma function 
    gamma_func = sp.gamma(α+β)/sp.gamma(α)*sp.gamma(β)
    
    #create variable to store probability of x given alpha and beta P(x|α,β)
    f_x_ab = Float64(gamma_func*(x^(α-1))*((1-x)^(β-1)))
end
```

### 1.2

1. Create a x from 0 to 1 with increment of 0.001 as the probability of event.

```julia
#1: Define probability of event x
x_range = 0.0:0.001:1.0
collect(x_range)
```

```julia
#2 Define values of alpha and beta
α = 0.5
β = 0.5
```

```julia
#Calculate the pdf of beta distribution over x using the pdf_beta function 
pdf_list = []
for i in 1:length(x_range)
    prob_i = pdf_beta(α,β,x_range[i])
    push!(pdf_list,prob_i)
end
```

The beta distribution has a U shape, with two peaks on either end.This is like a binomial distribution because alpha and beta are 0.5.
