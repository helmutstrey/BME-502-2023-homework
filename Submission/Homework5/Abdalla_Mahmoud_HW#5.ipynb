```julia
import SpecialFunctions
import Plots
```

```julia
function pdf_beta(α,β,x)
    p1 = gamma(α.+β) / (gamma(α).*gamma(β))
    p2 = x.^(α.-1)
    p3 = (1 .-x).^(β.-1)
pdf = p1.*p2.*p3

end
```

```julia
x = 0:0.001:1;
α = 0.5;
β = 0.5;
pdf = pdf_beta(x,α,β)
plot(x, pdf)
```

```julia
##The shape is a large U with a very wide base.
```

```julia
x = 0:0.001:1;
α = 1;
β = 1;
pdf = pdf_beta(x,α,β)
plot(x, pdf)
```

```julia
##The shape is now just a line instead of a U
```

```julia
x = 0:0.001:1;
α = 5;
β = 5;
pdf = pdf_beta(x,α,β)
plot(x, pdf)
```

```julia
##It is now a nonlinear graph resembling a wave graph
```

```julia
pdf_p = pdf_beta(35+α,65-35+β,x)
plot(x,pdf)
plot!(x,pdf_p)
```

```julia
## The plot now is a gaussian curve. 
## The peak got much smaller and moved to the right.
```
