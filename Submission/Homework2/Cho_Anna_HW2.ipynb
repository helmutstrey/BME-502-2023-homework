#### Bayes' theorem:
\begin{equation*}
P(A|B) = \frac{P(B|A)P(A)}{P(B)} \\
P(A|B): Posteior \\
P(B|A): Likelihood \\
P(A): Prior \\
P(B): Marginal Probability \\
\end{equation*}

The event is: The case of getting a positive test from the rapid antigen test

```julia
cov = 0.01
fcov = 1 - cov 
tpos = 0.653
fpos = 1 - tpos
marg = (tpos*cov)+(fpos*fcov)
println("Marginal Probability: Receiving positive test from rapid antigen test is $(marg*100)%.")
```

#### Define all the information and calculate the probability the randomly selected person has Covid

```julia
like = 0.653
posterior = (like*cov)/marg
println("The probability of the randomly selected person having COVID is $(posterior*100)%.")
```

### Read the following example and complete the exercises 
Bayes is playing Russian roulette. The revolver has six chambers. He puts two bullets in two adjacent chambers, spin the cylinder, hold the gun to his head, and pull the trigger. It clicks. So it is now the second try: he can choose either to spin the cylinder again or leave it as it is. 

### 4.1

#### What is the probability that Bayes survives the second try given his survival in the first try if he spin the cylinder and explain

The probability that Bayes survives the second time given he survived the first time is going to be the same as the probability of him surviving the first time. This is because he decided to spin the cylinder on the second try, which resets the chances of him surviving which is 4 out of 6 bullets, or 2/3 (66.7%).

### 4.2
#### In this case, what are the relation between P(B|A), P(B), and P(A) if A represents survival in the first try and B represents survival in the second try

Since Bayes spun the cylinder, the probability of him surviving the first (P(A)) and the second time (P(B)) are the same as the spin reset the chances. This means that P(A) is independent from P(B).

In the case of him surviving the second time given surviving the first time (P(B|A)), we would need to follow the formula of P(AnB)/P(A). 

Knowing that P(A) and P(B) are independent, we can simplify the formula to:
P(B|A) = P(A)P(B)/P(A) = P(B) = 2/3 (66.7%)

### 4.3

#### What is the probability that Bayes survives both tries without spinning the cylinder and explain

The probability of Bayes surviving both tries without spinning the cylinder is going to be when there are two consecutive blanks in the cylinder. 

There are 3 positions out of the total 6 positions that Bayes could be in if he survived the first and second time without spinning the cylinder.

The probability of Bayes surviving both tries without spinning the cylinder is 3/6, or 1/2 (50%).

### 4.4

#### Use answers from 4.1 and 4.3 to calculate the probability that Bayes survives the second try given his survial in the first try if he doesn't spin the cylinder and explain 

In the case of Bayes surviving the second time given surviving the first time (P(B|A)) without spinning the cylinder, we would need to follow the formula of P(AnB)/P(A).

From 4.1, we find that the chances of him surviving the first time is 2/3.

From 4.3, we find that the chances of him surviving the first and second time without spinning the cylinder is 1/2.

P(B|A) = P(AnB)/P(A) = (1/2) / (2/3) = 3/4
The probability of Bayes surviving the second time given surviving the first time without spinning the cylinder is 3/4 (75%).
