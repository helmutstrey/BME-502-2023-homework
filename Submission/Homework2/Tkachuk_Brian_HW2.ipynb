#### Bayes' theorem:
\begin{equation*}
P(A|B) = \frac{P(B|A)P(A)}{P(B)} \\
P(A|B): Posteior \\
P(B|A): Likelihood \\
P(A): Prior \\
P(B): Marginal Probability \\
\end{equation*}

1. Past data tells you that 10% of patients entering the clinic have liver disease (**Prior**). P(A)

2. Five percent of the clinic’s patients are alcoholics (**Marginal Probability**). P(B)

3. Among the patients diagnosed with liver disease, 15% are alcoholics (**Likelihood**). P(B|A)

Now you wonder what is the probability of a patient from the clinic have liver disease given that the patient is alcoholic (**Posterior**) P(A|B)
