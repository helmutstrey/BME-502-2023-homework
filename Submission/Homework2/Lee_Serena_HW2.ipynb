The event is: Positive Rapid Antigen Test

```julia
#P(P|C) = true positive
#P(P|NC) = false positive

#P(P) = {P(P|C) * P(C)} + {P(P|NC) * P(NC)}
#P(P) = (0.653 * 0.01) + ((1-0.653) * 0.99)
marginalProbability = (0.653 * 0.01) + ((1-0.653) * 0.99)
```

Bayes is playing Russian roulette. The revolver has six chambers. He puts two bullets in two adjacent chambers, spin the cylinder, hold the gun to his head, and pull the trigger. It clicks. So it is now the second try: he can choose either to spin the cylinder again or leave it as it is. 

### 4.1

#### What is the probability that Bayes survives the second try given his survival in the first try if he spin the cylinder and explain

 The probability that Bayes survives the second try given his survival in the first try if he spins the cylinder is 4/6 or 2/3. This is because there is still 4 chamber positions that are cleared that would allow Bayes to survive the second try. This translates to a 4/6 chance that he survives the second try when spinning the cylinder after the first try.

# 4.2
#### In this case, what are the relation between P(B|A), P(B), and P(A) if A represents survival in the first try and B represents survival in the second try

A = survival in first try ; B = survival in second try; P(B|A) = Probability of survival in second try given survival in first try

P(A) and P(B) are the same in the case when Bayes spins the cylinder. Using this and the conditional probability relationship P(B|A) = P(AB)/P(B) , P(B|A) will also be 2/3.

### 4.3

#### What is the probability that Bayes survives both tries without spinning the cylinder and explain

The probability that Bayes survives the second try given his survival in the first try if he doesn't spin again is 1/2. This is because his survival in the first try indicates that there are only 4 chamber positions that would allow Bayes to survive the first shot. Of these 4 positions, 3 will allow him to survive the second shot if he were not to spin the cylinder again. This results in a 3/6 or 1/2 chance of survival.

Using the equation P(B|A) = P(AB)/P(B), the probability that Bayes survives the second try given his survival in the first try when he doesn't spin the cylinder can be calculated. P(AB) is 3/6 or 1/2 because there are 3 of 6 total positions that are consecutively not loaded. P(A) is 4/6  since of the 4 of 6 total positions will be unloaded. Therefore, P(B|A) = (1/2)(2/3) = 0.75.
