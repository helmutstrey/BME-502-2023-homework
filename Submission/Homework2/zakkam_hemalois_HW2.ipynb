#### Bayes' theorem:
\begin{equation*}
P(A|B) = \frac{P(B|A)P(A)}{P(B)} \\
P(A|B): Posteior \\
P(B|A): Likelihood \\
P(A): Prior \\
P(B): Marginal Probability \\
\end{equation*}

#### What is the probability that Bayes survives the second try given his survival in the first try if he spin the cylinder and explain

The probability of the survivval in the second try depending on the survival in the first try would be 4/6 since the chances of survival are 4 times out of 6 possibilities.
