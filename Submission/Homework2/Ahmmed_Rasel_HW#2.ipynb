# 2.1
The probability of which event is considered marginal probability in this case? 

Define it in the first cell below and calculate it in the second cell below

The event is: Covid with test positive 

```julia
# P(test_pos)=P(test_pos|covid)*P(covid)+P(test_pos|neg_covid)*(1-P(covid))
Prob_test_pos_given_covid=0.653
Prob_covid=0.01
Prob_test_pos_given_neg_covid=1-0.999
Prob_test_pos=((Prob_test_pos_given_covid*Prob_covid)+(Prob_test_pos_given_neg_covid*(1-Prob_covid)))
```

```julia
# P(test_pos)=P(test_pos|covid)*P(covid)+P(test_pos|neg_covid)*(1-P(covid))=0.007520000000000001
# P(test_pos|covid)=0.653
# P(covid)=0.01
# P_covid_given_test_pos=P(test_pos|covid)*P(covid)/P(test_pos)
Prob_covid_given_test_pos=(Prob_test_pos_given_covid*Prob_covid)/Prob_test_pos
```

Read the following example and complete the exercises
Bayes is playing Russian roulette. The revolver has six chambers. He puts two bullets in two adjacent chambers, spin the cylinder, hold the gun to his head, and pull the trigger. It clicks. So it is now the second try: he can choose either to spin the cylinder again or leave it as it is.



# 4.1
What is the probability that Bayes survives the second try given his survival in the first try if
he spin the cylinder and explain


```julia
# Let A:{Survival on the 1st try}
# B:{Survival on the 2nd try}
# Prob_surv_2nd_given_surv_1st=P(2nd|1st)
# Prob_surv_2nd_given_surv_1st=Prob_surv_1st_given_surv_2nd

# Spinning means that we rotate the barrel first, it erases history and
# makes the second shot identical to the first one.Thus, if we spin the barrel,
number_of_empty_chamber_after_1st_try=4
Total_number_of_chamber=6
Prob_B_given_A = (number_of_empty_chamber_after_1st_try/Total_number_of_chamber)  
```

# 4.2
In this case, what are the relation between P(B|A), P(B), and P(A) 
if A represents survival in the first try and B represents survival in the second try


P(B|A) equal to P(A) because there are two chambers filled with bullets, out of six chambers as it erases the history 
and they are independent.

# 4.3
What is the probability that Bayes survives both tries without spinning the cylinder and explain


```julia
#Let A:{Survival on the 1st try}
# B:{Survival on the 2nd try} 
# However, if we do not spin, the second shot is not identical to the first. In fact 
number_of_empty_chamber_after_1st_try=3
Total_number_of_chamber=6
Prob_surv_both_A_and_B=(number_of_empty_chamber_after_1st_try/Total_number_of_chamber)
# because there are three chambers such that both chamber and the one next to it are empty.
```

# 4.4
Use answers from 4.1 and 4.3 to calculate the probability that Bayes survives the second try given 
his survial in the first try if he doesn't spin the cylinder and explain

```julia
# Let A:{Survival on the 1st try}
# B:{Survival on the 2nd try} 
# So, if we do not spin,
Prob_B_given_A= Prob_surv_both_A_and_B/Prob_B_given_A 
# Hence Bayes should ask to pull the trigger again without randomly spinning 
# the barrel first to maximize the chance of his survival.
```
