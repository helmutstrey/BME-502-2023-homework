#### Bayes' theorem:
\begin{equation*}
P(A|B) = \frac{P(B|A)P(A)}{P(B)} \\
P(A|B): Posteior \\
P(B|A): Likelihood \\
P(A): Prior \\
P(B): Marginal Probability \\
\end{equation*}

# Question 1
### Read the following example and complete the exercises 

You are at the age of 21 and curious about if drinking alcohol would cause liver disease. You contacted a clinic site and got the following information.

1. Past data tells you that 10% of patients entering the clinic have liver disease (**Prior**). 

2. Five percent of the clinic’s patients are alcoholics (**Marginal Probability**). 

3. Among the patients diagnosed with liver disease, 15% are alcoholics (**Likelihood**). 

Now you wonder what is the probability of a patient from the clinic have liver disease given that the patient is alcoholic (**Posterior**)

### 1.1 

#### If A represents the event of being an alcoholic patient in the clinic and L represents the event of being a patient with liver disease in the clinic, write the bayesian representation (P(A|B)) in the following cells


Posteria = P(L|A)

Prior = P(L)

Likelihood = P(A|L)

Marginal Probability = P(A)

### 1.2 

#### Define all the information available and calculate the posteria probability

<b>Posteria</b>, P(L|A), is the probability of a patient having liver disease GIVEN they are alcoholic. Determined in next cell...

<b>Prior</b>, P(L), is the probability of a patient in the clinic having liver disease. 10%.

<b>Likelihood</b>, P(A|L), is the probability of a patient in the clinic being an alcoholic GIVEN they have liver disease. 15%.

<b>Marginal Probability</b>, P(A), is the probability of a patient in the clinic being alcoholics. 5%.

```julia
## Calculate posteria probability

# Initialize knowns
prior = 0.1
likelihood = 0.15
marginal_probability = 0.05

# Calculate posterior probability
posterior = (likelihood * prior) / marginal_probability
```

Assume 1% people have Covid. If randomly select a person and he/she tests positive using the rapid antigen test, what is the probability the person has Covid (posterior).

### 2.1 

#### The probability of which event is considered marginal probability in this case? Define it in the first cell below and calculate it in the second cell below

The event is: <b>Probability of positive result using the rapid antigen test</b>

```julia
sensitivity = 0.653 # probability of a positive test result given the patient is infected (likelihood)
1-sensitivity # probability of a negative test result given the patient is infected

specificity = 0.999 # probability of a negative test result given the patient is not infected
1-specificity # probability of a positive test result given the patient is not infected

prevalence = 0.01 # percent of people having COVID (prior)

# marginal_probability calculation
marginal_probability = (prevalence*sensitivity)+((1-specificity)*(1-prevalence))
```

### 2.2 

#### Define all the information and calculate the probability the randomly selected person has Covid

```julia
likelihood = 0.653 # probability of a positive test result given the patient is infected (sensitivity)
prior = 0.01 # probability of a patient infected with COVID
marginal_probability = 0.00752 # probability of a positive test result

# Posterior calculation
posterior = (likelihood * prior) / (marginal_probability) # probability of a patient infected with COVID given a positive test result
```

### Read the following example and complete the exercises 
Bayes is playing Russian roulette. The revolver has six chambers. He puts two bullets in two adjacent chambers, spin the cylinder, hold the gun to his head, and pull the trigger. It clicks. So it is now the second try: he can choose either to spin the cylinder again or leave it as it is. 

### 4.1

#### What is the probability that Bayes survives the second try given his survival in the first try if he spins the cylinder and explain

There is always a 4/6 chance of survival after spinning the cylinder since there are two bullets.<br>

P(A) = 4/6 # First try survival chance<br>
P(B) = 4/6 # Second try survival chance<br>
...<br>
P(Z) = 4/6 # Z try survival chance<br><br>

We may express the conditional representation of second try survival chance given the first below.<br>
P(B|A) = P(B)*P(A|B)/P(A) = P(B) # Second try survival chance given first try survival<br>

However, P(A) and P(B) are independent events since spinning the cylinder removes the conditional aspect.<br>
<b>P(B|A) = P(B) = 4/6 = 2/3</b> # Second try survival chance given first try survival.<br><br>

### 4.2
#### In this case, what are the relation between P(B|A), P(B), and P(A) if A represents survival in the first try and B represents survival in the second try

P(B|A) = P(B)*P(A|B)/P(A) # Conditional probability of B given A<br>
P(A|B) = P(A)*P(B|A)/P(B) # Conditional probability of A given B<br>

Note if P(A) and P(B) are <b>independent</b> from one another (spinning the cylinder), then conditional probability of B given A, or A given B, is P(B) and P(A) respectively.<br><br>

### 4.3

#### What is the probability that Bayes survives both tries without spinning the cylinder and explain

P(A) = 4/6 # Probability of surviving first try.<br>

After the shot is fired, P(B) = 3/4 since there will be a single bullet in one of the next 4 chambers.<br>

Multiplying these probabilities give the overall chance of survival for first and second night.<br>
P(A and B) = 4/6 * 3/4 = <b>50% chance of survival for both tries without spinning.</b>
