#### Bayes' theorem:
\begin{equation*}
P(A|B) = \frac{P(B|A)P(A)}{P(B)} \\
P(A|B): Posteior \\
P(B|A): Likelihood \\
P(A): Prior \\
P(B): Marginal Probability \\
\end{equation*}

### Read the following example and complete the exercises 
Shown by a <a href="https://pubmed.ncbi.nlm.nih.gov/34242764/#:~:text=The%20overall%20sensitivity%20of%20the,%25%20CI%2024.4%2D65.1">reaserch study</a>, the rapid antigen test for Covid has a sensitivity of 65.3% and a specificity of 99.9%. Assume 1% people have Covid. If randomly select a person and he/she tests positive using the rapid antigen test, what is the probability the person has Covid (posterior).

### 2.1 

#### The probability of which event is considered marginal probability in this case? Define it in the first cell below and calculate it in the second cell below

The event is: The event that person tests positive using the rapid antigen test P(P)

```julia
# sensitivity - ability of the test to identify the presence of a disease or illness correctly (probability of a true positive)
# specificity - ability of the test to identify the absence of a disease or illness correctly (probability of a true negative)
# C = the event that a person has COVID
# P = the event that someone tests positive 
# N = the event that someone tests negative
# NC = the event that someone does not have COVID (healthy)
P_C = 0.01 # P(C) the probability that someone has COVID 
P_NC = 1 - P_C # P(NC) the probability that someone does not have COVID/healthy
P_PC = 0.653 # P(P|C) the probability that someone tests positive given that they have COVID
P_N_C = 1 - P_PC # P(N|C) the probability that someone tests negative given that they have COVID
P_NNC = 0.999 # P(N|NC) the probability that someone test negative given they do not have COVID
P_PNC = 1 - P_NNC # P(P|NC) the probability that someone tests positive given they do not have COVID
P_PandC = P_PC*P_C # P(P∩C) the probability that someone tests positive and they have COVID
P_PandNC = P_PNC*P_NC # P(P∩NC) the probability that someone tests positive and they do not have COVID
P_P = (P_PandC) + (P_PandNC) # P(P) the probability that someone tests positive
```

### 2.2 

#### Define all the information and calculate the probability the randomly selected person has Covid

```julia
# P_C_random - probability that a randomly selected person has COVID
# P_PC - probability that a person who tests positive, given they have COVID
# P_C - probability that a person has COVID
# P_P - probability that a person tests positive

P_C_random = (P_PC*P_C)/P_P
```

### 4.1

### 4.2
#### In this case, what are the relation between P(B|A), P(B), and P(A) if A represents survival in the first try and B represents survival in the second try

P(A) is the probability that he survives the first shot (4/6)
P(B) is the probability that he survives the second shot (4/6)
P(B|A) is the probability that he survives the second shot given he also survives the first shot

We know that Bayes takes a shot and survives. This is the success of event A. He then spins the cylinder and takes another shot. Since B is independent of A, they have the same probabilities. As a result, P(B|A) is also 4/6 or 2/3. Essentially, Bayes resets the probabilities by spinning the chamber, meaning his second shot is probabilistically the same as the first.

This is because we can eliminate the possibility of the chamber being in the position of XXOOOO and XOOOOX, where X represents a bullet and O represents an empty slot. This leaves us with 4 possibilties remaining, only one of which may have a bullet in the upcoming chamber.
