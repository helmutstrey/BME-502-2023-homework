#### Bayes' theorem:
\begin{equation*}
P(A|B) = \frac{P(B|A)P(A)}{P(B)} \\
P(A|B): Posterior \\
P(B|A): Likelihood \\
P(A): Prior \\
P(B): Marginal Probability \\
\end{equation*}

### Read the following example and complete the exercises 
Shown by a <a href="https://pubmed.ncbi.nlm.nih.gov/34242764/#:~:text=The%20overall%20sensitivity%20of%20the,%25%20CI%2024.4%2D65.1">reaserch study</a>, the rapid antigen test for Covid has a sensitivity of 65.3% and a specificity of 99.9%. Assume 1% people have Covid. If randomly select a person and he/she tests positive using the rapid antigen test, what is the probability the person has Covid (posterior).

### 2.1 

#### The probability of which event is considered marginal probability in this case? Define it in the first cell below and calculate it in the second cell below

The event is: 

```julia
#Marginal Probability = P(test positive) = P(test positive|has covid)*P(has covid) + P(test positive|no covid)*P(no covid)
#Prior = P(has covid) = 0.01
#Likelihood = P(test positive|has covid) = 0.653
#Posterior = P(has covid|test positive) = {P(test positive|has covid)*P(has covid)}/P(test positive)
#P(test positive|no covid) = 1 - Cov_specificity ; the chances of a false positive
P_hasCov = 0.01
P_noCov = 0.99 # 1 - P_noCov
Cov_likelihood = 0.653 # the capabilities/chances of the test identifying covid
Cov_specificity = 0.999 # the accuracy of the test; inverse is chances of false positive {P(test positive|no covid)}
P_testPos = (Cov_likelihood*P_hasCov)+((1-Cov_specificity)*P_noCov)
P_testPos_percent = (P_testPos * 100)
println("The marginal probability is the probability of testing positive, which is $P_testPos or $P_testPos_percent percent")
```

### 2.2 

#### Define all the information and calculate the probability the randomly selected person has Covid

```julia
P_hasCov = 0.01
P_noCov = 0.99
Cov_likelihood = 0.653
Cov_specificity = 0.999
# P(test positive|no covid) = 1 - Cov_specificity
P_testPos = (Cov_likelihood*P_hasCov)+((1-Cov_specificity)*P_noCov)
Cov_posterior = (Cov_likelihood*P_hasCov)/P_testPos
Cov_posterior_percent = (Cov_posterior * 100)
println("The probability of a randomly seleced person having Covid after testing positive is $Cov_posterior or $Cov_posterior_percent percent")
```

### 4.2
#### In this case, what are the relation between P(B|A), P(B), and P(A) if A represents survival in the first try and B represents survival in the second try

P(A) = first round survival
P(B) = second round survival
P(A∩B) = surviving the first and second round
P(B|A) = survive second round given survived first round
P(B|A) = P(A∩B)/P(A)
