```julia
import Pkg
Pkg.add(["Distributions", "Plots", "Polynomials", "Random"])
```

```julia
import Distributions as di
import Plots
import Polynomials
import Random
```

```julia
x = -10:1:10
y_true = (x.^2) .+ (3 .* x) .- 2
```

```julia
dataerr = 10
noise = di.Normal(0,dataerr)
y_observed = y_true + rand(noise, length(x))
```

```julia
using Plots
scatter(x, y_observed)
plot!(x, y_true)
```

```julia
poly_fit= fit(x,y_observed,2)
y_fitted=poly_fit.(x) #fit_coeff[1] .+ fit_coeff[2] .*x .+ fit_coeff[3] .*x.^2.0
plot!(x,y_fitted, label="Fitted line")
```

```julia
dataerr = 50
noise = di.Normal(0,dataerr)
y_observed = y_true + rand(noise, length(x))
using Plots
scatter(x, y_observed)
plot!(x, y_true)
poly_fit= fit(x,y_observed,2)
y_fitted=poly_fit.(x) 
plot!(x,y_fitted, label="Fitted line")
```

```julia
# Due to the increase in standard 
# deviation the fitted line is not as close as it is in 2.3.
```

```julia
scatter(x,y_observed)
plot!(x,y_true)
plot!(x,fit(x,y_observed,10).(x))
```

```julia
x_fit = -11:.1:11
collect(x_fit)
scatter(x,y_observed)
plot!(x,y_true)
plot!(x_fit,fit(x,y_observed,10).(x_fit))
```

```julia
# We can see that y3(fitted) is attempting to be in between as many 
# points of data as possible and the line is skewed.
# This is known as overfitting.
```
