```julia
import Pkg
Pkg.add("Distributions")
```

```julia
function sample_gaussian(mu,sigma,n)
    gaus = zeros(n)
    for x in 1:n
        num = rand()
        gaus[x] = mu .+ (sigma .* sqrt(2) .* (erfinv((2 .* num) .- 1)))
    end
    gaus
end
```

```julia
using SpecialFunctions
using Plots
```

```julia
data = sample_gaussian(-1 , 2, 10000)
```

```julia
histogram(data, normalize=true)
```

```julia
using Distributions
x=-8:0.001:6
gauss =  pdf.(Normal(-1,2),x)
histogram(data,normalize=true)
plot!(x,gauss)
```
